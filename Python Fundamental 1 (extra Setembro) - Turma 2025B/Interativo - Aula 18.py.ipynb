```python
turno = input('Digite m para manhã, t para tarde, ou n para noite: ')
if turno =='m':
         print('Bom dia!')
elif turno =='t':
         print('Boa tarde!')
elif turno =='n':
         print('Boa noite!')
else:
         print('Letra inválida')
```
